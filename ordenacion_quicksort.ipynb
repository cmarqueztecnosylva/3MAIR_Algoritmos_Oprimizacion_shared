```
##########################################################################
# Quicksort
#
# https://www.interviewbit.com/tutorial/quicksort-algorithm/
# https://deepai.org/machine-learning-glossary-and-terms/quicksort-algorithm
##########################################################################
import numpy as np
import time

##########################################################################
# Metodo que forma parte de la ordenacion quicksort, y que hace ordenaciones
# parciales y devuelve el inciden del pivot en ltima instancia
##########################################################################
def divide_array(my_list, min_index, max_index):
  i_pivot = min_index - 1
  pivot = my_list[max_index]

  for j in range(min_index, max_index):
    if my_list[j] < pivot:
      i_pivot = i_pivot + 1
      
      # Intercambio
      my_list[i_pivot],my_list[j] = my_list[j],my_list[i_pivot]    
    #Endif
  #Endfor

  #
  my_list[i_pivot+1],my_list[max_index] = my_list[max_index],my_list[i_pivot+1] 
  return ( i_pivot+1 ) 


##########################################################################
# Mtodo de ordenacin Quicksort
##########################################################################
def my_quick_sort(my_list, min_index, max_index):
  
  #print('DEBUG: my_list ',my_list)

  if min_index < max_index: 
  
        # split_from es el indice que va a dividir el array en las
        # siguientes iteracioes
        split_from = divide_array(my_list,min_index,max_index) 

        #print('DEBUG: split_from ',split_from)

        # procesamos el array en dos partes diferentes mediante recursividad
        my_quick_sort(my_list, min_index, split_from-1) 
        my_quick_sort(my_list, split_from+1, max_index) 

##########################################################################
#
##########################################################################
#my_list = [9,6,7,3,1,12]
my_list = np.random.randint(1, 100, size=(500), dtype=int)

print('Lista de entrada:')
print(my_list)

#Tiempo inicial
t0 = time.time()

#Llamada a la ordenacin
my_quick_sort(my_list, 0, len(my_list) - 1)

#Tiempo final
t1 = time.time()
total_n = t1-t0

print('Lista de salida:')
print(my_list)
print('en un tiempo total de: ',total_n)
```
